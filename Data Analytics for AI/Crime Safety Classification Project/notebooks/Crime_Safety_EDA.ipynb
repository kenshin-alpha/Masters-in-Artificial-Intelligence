# Crime Safety Classification Project
## Exploratory Data Analysis (EDA)
This notebook explores the crime dataset and visualizes key patterns relevant to public space safety.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('../data/RCD06.20251204131643.csv')
df.head()
```

```python
# Example: Visualize crime counts by location
if 'Garda Region' in df.columns:
    plt.figure(figsize=(12,6))
    sns.countplot(data=df, x='Garda Region', order=df['Garda Region'].value_counts().index)
    plt.title('Crime Counts by Location')
    plt.xticks(rotation=90)
    plt.show()
```

```python
# Example: Visualize severity distribution
if 'severity' in df.columns:
    plt.figure(figsize=(8,5))
    sns.boxplot(data=df, x='severity')
    plt.title('Crime Severity Distribution')
    plt.show()
```
