```python
from amplpy import AMPL, ampl_notebook

ampl = ampl_notebook(
    modules=["cbc", "highs"],  # modules to install
    license_uuid="default",  # license to use
)  # instantiate AMPL object and register magics
```

```python
%%ampl_eval

var apples >= 0; 
var bananas >= 0; 
```

```python
%%ampl_eval
maximize Total_Fruit:
    apples + bananas;
subject to Banana_Limit:
    bananas <= 4;
subject to Apple_Limit:
    apples <= 9;
```

```python
%%ampl_eval
subject to Banana_Upper_Ratio:
    bananas <= 2 * apples;

subject to Banana_Lower_Ratio:
    bananas >= 0.5 * apples;
```

```python
ampl.eval("show;")
ampl.eval("expand;")
```

```python
ampl.option["solver"] = "cbc"
ampl.solve()
```

```python
ampl.option["solver"] = "highs"
ampl.solve()
```
